```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Biological Setup
file = 'Seq_information.txt'

# Challenge
# Open the file in pandas
# Show its first 5 and bottom 5 rows
# Print its column
# Print the number of rows
```

```python
df = pd.read_csv('Seq_information.txt', sep='\t')
print(df.head())
print("-------------------------------------------------------------------------")
print(df.tail())
print("-------------------------------------------------------------------------")
print(df.columns)
print("-------------------------------------------------------------------------")
print(len(df))
```

```python
# Challenge
# Is there any association between GC content vs Sequence Length
# Visualise the Sequence Length Distribution
# Visualise the GC Skew Distribution
# Use heatmaps to show the correlation of various features
# categorize sequences by GC Content:
```

```python
# Scatter Plot: GC Content vs Sequence Length
sns.scatterplot(x='Length', y='GC Content', data=df)
plt.title('GC Content vs Sequence Length')
plt.xlabel('Sequence Length (bp)')
plt.ylabel('GC Content')
plt.grid(True)
plt.show()
```

```python
# Histogram: Sequence Length Distribution
sns.histplot(df['Length'], bins=20, kde=True)
plt.title('Length Distribution')
plt.xlabel('Length')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Histogram: GC Skew Distribution
sns.histplot(df['GC Skew'], bins=20, kde=True)
plt.title('GC Skew Distribution')
plt.xlabel('GC Skew')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
df['GC_category'] = pd.cut(df['GC Content'], bins=[0, 0.4, 0.45, 1], labels=['Low', 'Medium', 'High'])
df.head()
```

```python
df['GC_category'].value_counts().plot.pie(autopct='%1.1f%%', shadow=True,
                                          explode = (0.0, 0.0, 0.3))
plt.show()
```
